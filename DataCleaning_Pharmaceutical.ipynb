#### Normalizing active_ingredients column

```python
print(df.loc[df['num_active_ingredients']>2])

df.drop(df.loc[df['num_active_ingredients']>2].index, inplace=True)

```

```python
print(df.loc[df['num_active_ingredients']>2])
```

```python
import ast
df['active_ingredients'] = df['active_ingredients'].apply(ast.literal_eval)
```

```python
df['active_ingredients'].apply(type).value_counts()
```

```python
df['secondary_ingredient'] = df['active_ingredients'].apply(
    lambda x: x[1]['name'] if isinstance(x, list) and len(x) >= 2 else None
)
```

```python
df['secondary_strength'] = df['active_ingredients'].apply(
    lambda x: x[1]['strength'] if isinstance(x, list) and len(x) >= 2 else None
)
```

```python
df.drop(columns='active_ingredients', inplace=True)
```

#### Check whether drug is combination or not


```python
df['combination_drug'] = df['num_active_ingredients'].apply(lambda x: True if x > 1 else False)
```

#### Check manufacturer & manufacturer_raw column similarity

```python
df['manufacturer_same'] = df['manufacturer_raw'] == df['manufacturer']
df['manufacturer_same'].value_counts()
```

```python
df.loc[df['manufacturer_same'] == False,
       ['manufacturer_raw', 'manufacturer']]
```

```python
df.drop(columns = ['manufacturer_raw','manufacturer_same'], inplace=True)
```
